```python
# разные режимы датасета
DATA_MODES = ['train', 'val', 'test']
# все изображения будут масштабированы к размеру 224x224 px
RESCALE_SIZE = 224
BATCH_SIZE = 128
N_CLASSES = 42
EPOCHS = 30
```

```python
class SimpsonsDataset(Dataset):
    """
    Датасет с картинками, который паралельно подгружает их из папок
    производит скалирование и превращение в торчевые тензоры
    """
    def __init__(self, files, mode):
        super().__init__()
        # список файлов для загрузки
        self.files = sorted(files)
        # режим работы
        self.mode = mode

        if self.mode not in DATA_MODES:
            print(f"{self.mode} is not correct; correct modes: {DATA_MODES}")
            raise NameError

        self.len_ = len(self.files)

        self.label_encoder = LabelEncoder()

        if self.mode != 'test':
            self.labels = [path.parent.name for path in self.files]
            self.label_encoder.fit(self.labels)

            with open('label_encoder.pkl', 'wb') as le_dump_file:
                  pickle.dump(self.label_encoder, le_dump_file)

    def __len__(self):
        return self.len_

    def load_sample(self, file):
        image = Image.open(file)
        image.load()
        return image

    def __getitem__(self, index):
        # Здесь на радномно происходит одно преобразование
        # Где-то настроил вероятности так, чтобы не все изображения на трейне были изменены
        augmentations = transforms.RandomChoice([
            transforms.Resize(size=(RESCALE_SIZE, RESCALE_SIZE)), # это тоже для отстутствия изменений
            transforms.Compose([
                transforms.Resize(size=300, max_size=301),
                transforms.CenterCrop(size=300),
                transforms.RandomCrop(250)
            ]),
            transforms.RandomRotation(degrees=(-45, 45)), # случайный поворот от -45 градусов до 45
            transforms.RandomHorizontalFlip(p=0.7), # горизонтальное отражение
            transforms.RandomAdjustSharpness(sharpness_factor=np.random.choice(np.arange(0.9, 1.2, 0.1)), p=1), # случайное изменение резкости
            transforms.RandomPosterize(bits=6) # случайное изменение глубины цвета
        ])
        normalize_transform = transforms.Compose([
            transforms.Resize(size=(RESCALE_SIZE, RESCALE_SIZE)),
            transforms.Lambda(lambda x: np.array(x, dtype="float32") / 255),
            # torch советует использовать следующую конструкцию вместо устаревшей ToTensor():
            transforms.Compose([transforms.ToImageTensor(), transforms.ConvertImageDtype()]), 
            transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
        ])

        x = self.load_sample(self.files[index])
        if self.mode == 'train':
            x = augmentations(x)
        x = normalize_transform(x)
        if self.mode == 'test':
            return x
        else:
            label = self.labels[index]
            label_id = self.label_encoder.transform([label])
            y = label_id.item()
            return x, y
```

```python
train_dataset = SimpsonsDataset(train_files, mode='train')
val_dataset = SimpsonsDataset(val_files, mode='val')
test_dataset = SimpsonsDataset(test_files, mode='test')
```

```python
def create_dct_path_labels(train_files, train_labels):
    dct_simpsons = {}
    for label_i in np.unique(train_labels).tolist():
        dct_simpsons[label_i] = []

    for path_i, label_i in zip(train_files, train_labels):
        dct_simpsons[label_i].append(path_i)

    return dct_simpsons


# Создадим словарь в котором ключами будут персонажи Симпсонов, а значениями списки с путями к картинкам.
dct_path_train = create_dct_path_labels(train_files, train_dataset.labels)
```

```python
# Дополним картинки классов у которых менее MIN_SIZE картинок, до MIN_SIZE картинок в классе
MIN_SIZE = 500
for person in dct_path_train:
    if len(dct_path_train[person]) < MIN_SIZE:
        dct_path_train[person] = dct_path_train[person] * \
            (MIN_SIZE // len(dct_path_train[person]))
        dct_path_train[person].extend(
            dct_path_train[person][:MIN_SIZE - len(dct_path_train[person])])
# Проверим что получилось
vis_dict = {}
for person in dct_path_train:
    vis_dict[person] = len(dct_path_train[person])
    # print(f"{person}\t{len(dct_path_train[person])}")
new_train_files = []

vis_series = pd.Series(vis_dict.values(), index=vis_dict.keys()).sort_values()

plot_classes(vis_series.index, 
            vis_series.values,
            "Распределение классов на аугментированном трейне",
            lines=[2, 5])

for person in dct_path_train:
    new_train_files.extend(dct_path_train[person])

new_train_dataset = SimpsonsDataset(new_train_files, mode='train')
```

```python
def imshow(inp, title=None, plt_ax=plt, default=False):
    """Imshow для тензоров"""
    inp = inp.numpy().transpose((1, 2, 0))
    mean = np.array([0.485, 0.456, 0.406])
    std = np.array([0.229, 0.224, 0.225])
    inp = std * inp + mean
    inp = np.clip(inp, 0, 1)
    plt_ax.imshow(inp)
    if title is not None:
        plt_ax.set_title(title)
    plt_ax.grid(False)
```

# Обучение моделей (PyTorch Lightning)

Начну с того, что лайтнинг переезжает с адреса pytorch_lightning на lightning, которого нет на Kaggle, но новую версию скачать можно.  Вообще, сейчас они представляют из себя копии друг друга, так что с этим можно не заморачиваться. Также новая документация расположена на lightning.ai, на который можно зайти только с ВПНом. 


В лайтнинге не нужно вручную создавать циклы эпох, достаточно создать класс, наследованный от LightningModule, в котором определены методы __init__, forward (не уверен здесь, что надо, может, он автоматически встает), training_step, validation_step. Также, если хотите, можно определить test_step (для вызова теста), оптимайзера (у меня в коде они есть). Потом создается объект Trainer, в который передается новый класс и у него есть много интересных параметров. Будем рассматривать все по порядку.

Также как заявлено в документации, он работает быстрее, чем чистый torch.

```python
class ImgClassifier(pl.LightningModule):

    def __init__(self, model, optimizer):
        super().__init__()
        self.model = model # модель - предзагруженная или созданная нами
        self.optimizer = optimizer # оптимайзер - создал для каждой модели своей (дальше в коде)
        self.loss = nn.CrossEntropyLoss() # лосс
        # Вот шедулер выбрал более интересный, чем StepLR. Этот уменьшает lr, когда заданная вами метрика
        # перестает возрастать кол-во=patience эпох. Более подробно можно посмотреть в документации
        self.scheduler = lr_scheduler.ReduceLROnPlateau( 
            self.optimizer,
            mode='min',
            factor=0.1,
            patience=2,
            threshold=0.01,
            min_lr=1e-7,
            verbose=True
        )
        # Также можно задать метрики. Они автоматически считаются в конце эпох, если их импортировать
        # из torchmetrics (либо создать свою на основе соответствующего класса из torchmetrics)
        # Стоит заметить, что метрики из sklearn не будут нормально работать, потому что в них
        # не реализована сложная логика и синхронизация между эпохами, как это сделано в torchmetrics
        self.f1_train = MCF1(N_CLASSES, average='micro')
        self.f1_val = MCF1(N_CLASSES, average='micro')

    def forward(self,x):
        return self.model(x)

    def training_step(self, tr_batch, batch_idx):
        # Training step - думаю, тут всё понятно
        
        images, labels = tr_batch
        predictions = self(images) # Пропускаем батч через себя (метод forward)

        loss = self.loss(predictions, labels)

        self.log(
            "train_loss",
            loss,
            on_step=False, on_epoch=True, prog_bar=True, logger=True
        )
        
        self.log(
            'train_f1',
            self.f1_train(predictions, labels),
            on_step=False, on_epoch=True, prog_bar=True, logger=True
        )

        return loss

    def validation_step(self, val_batch, batch_idx):
        # То же самое, только для валидации

        images, labels = val_batch
        
        
        # Для валидации MobileOne желательно вызывать reparameterize для оптимизации расхода памяти
        # и ускорения инференса за счет упрощения конструкций со skip-connectian'ами
        # (подробнее в статье на arXiv'е от авторов MobileOne, ссылка есть в их гитхабе)
        if str(self.model)[:9] == 'MobileOne':
            model = reparameterize_model(self.model)
            predictions = model(images)
        else:
            predictions = self(images)
        
        loss = self.loss(predictions, labels)

        self.log(
            "val_loss",
            loss,
            on_step=False, on_epoch=True, prog_bar=True, logger=True
        )
        
        self.log(
            'val_f1',
            self.f1_val(predictions, labels),
            on_step=False, on_epoch=True, prog_bar=True, logger=True
        )

        return loss
    
    def predict_step(self, batch, batch_idx):

        if str(self.model)[:9] == 'MobileOne':
            model = reparameterize_model(self.model)
            predictions = model(batch)
        else:
            predictions = self(batch)
        return predictions
        
    
    def configure_optimizers(self):
        # Здесь надо указать оптимайзеры и шедулеры, а так же основную метрику для мониторинга
        return {
            'optimizer': self.optimizer,
            'lr_scheduler': self.scheduler,
            'monitor': 'val_loss'
        }
```

```python
train_loader = DataLoader(new_train_dataset, batch_size=BATCH_SIZE, shuffle=True, num_workers=N_CORES)
val_loader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=N_CORES)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False, num_workers=N_CORES)
```

## MobileOne S0

Сначала используем модельку MobileOne-S0, обучим ее полностью на одном learning rate.

```python
mobile_s0 = mobileone(variant='s0')

m_checkpoint = torch.load('/kaggle/input/mobileone-s0/mobileone_s0_unfused.pth.tar')
mobile_s0.load_state_dict(m_checkpoint)
```

```python
#Wandb используется для отслеживания кучи разных метрик, подробнее на их сайте
wandb_logger_mobile = WandbLogger(
    project='Journey to Springfiled',
    log_model="all"
)

# Также тут настроим кастомные чекпоинты
checkpoint_callback_mobile = ModelCheckpoint(
    monitor="val_loss",
    mode="min",
    dirpath='/kaggle/working/mobile_checkpoints',
    filename='{epoch}-{val_loss:.3f}-{val_f1:.3f}',
    save_top_k=1
)

# EarlyStopping по аналогии с бустингами
early_stopping_mobile = EarlyStopping( 
    monitor='val_loss',
    mode='min', # min - значит метрика должна падать, он среагирует если она этого делать не будет
    patience=4,
    min_delta=0.00,
    verbose=True
)
```

```python
mobile_s0.linear = nn.Linear(in_features=1024, out_features=N_CLASSES, bias=True) # меняем последний слой
lr=1e-4
optimizer = AdamW(mobile_s0.parameters(), lr=lr) # задаем оптимайзер

clf_mobile = ImgClassifier(mobile_s0, optimizer=optimizer).to(device) # создаем класс

# В конфигурацию обучения можно добавить информацию
wandb_logger_mobile.experiment.config.update({
    'batch_size': BATCH_SIZE,
    'max_epochs': EPOCHS,
    'early_stopping': 'yes',
    'scheduler': 'plain',
    'lr_start': lr,
    'training_type': 'all layers, same lr',
    'test_run': 'yes' if EPOCHS < 10 else 'no'
})

trainer_mobile = pl.Trainer(
    max_epochs=EPOCHS, # максимальное кол-во эпох (можно задать много, т. к. есть EarlyStopping и чекпоинты)
    enable_checkpointing=True, # чекпоинтинг и так включен по умолчанию, здесь просто показал
    callbacks=[
        early_stopping_mobile,
        checkpoint_callback_mobile
    ],
    accelerator="gpu",
    logger=wandb_logger_mobile
)

# Одной строчкой запускаем обучение модели
trainer_mobile.fit(model=clf_mobile, train_dataloaders=train_loader, val_dataloaders=val_loader)
```

# Предсказания и сабмит

```python
def get_int_preds(preds):
    preds_max = torch.concatenate(preds).numpy()
    preds_max = np.argmax(preds_max, axis=1)
    preds_max = preds_max.astype(int)
    return preds_max

def get_probs(preds):
    preds_probs = torch.concatenate(preds)
    return softmax(preds_probs, dim=1).numpy()
    
def make_submission_file(preds, name):
    preds_names = train_dataset.label_encoder.inverse_transform(preds)
    test_img_names = [path.name for path in test_files]
    sub = pd.DataFrame({'Id': test_img_names, 'Expected': preds_names})
    sub['sorting'] = sub['Id'].apply(lambda x: int(x.replace('img', '').replace('.jpg', '')))
    sub = sub.sort_values('sorting').reset_index(drop=True)
    sub.drop(columns='sorting', inplace=True)
    sub.to_csv(f'/kaggle/working/{name}.csv', index=False)
```

```python
# Здесь можно просто указать, что надо использовать лучший чекпоинт
preds_raw_mobile = trainer_mobile.predict(ckpt_path="best", dataloaders=test_loader)

preds_int_mobile = get_int_preds(preds_raw_mobile)
preds_probs_mobile = get_probs(preds_raw_mobile)

make_submission_file(preds_int_mobile, 'mobile_s0_20_epochs')
```

```python
preds_raw_tv = trainer_tinyvit.predict(ckpt_path="best", dataloaders=test_loader)

preds_int_tv = get_int_preds(preds_raw_tv)
preds_probs_tv= get_probs(preds_raw_tv)

make_submission_file(preds_int_tv, 'tiny_vit_25_epochs')
```

```python
# Тут сблендил по приколу вероятности классов с коэффициентами 0.8 и 0.2
# Разница такая большая, т. к. TinyVit модель намного точнее, подбирал на глаз
f_k, s_k = 0.8, 0.2
probs_blended = (f_k * preds_probs_tv) + (s_k * preds_probs_mobile)
preds_blended = np.argmax(probs_blended, axis=1)

make_submission_file(preds_blended, f'{f_k}tv_{s_k}mobile')
```
